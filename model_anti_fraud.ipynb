```python
# load data
import pandas as pd

dt = pd.read_csv("/content/gdrive/MyDrive/data/creditcard.csv")
dt.head()
```

```python
# đếm số dòng
dt.count()
```

```python
# số dòng dữ liệu trong mỗi class
dt['Class'].value_counts()
```

```python
# Drop cột time
dt_no_time = dt.drop(columns=('Time'))
dt_no_time.head()
```

```python
# Chuẩn hóa lại cột Amount
from sklearn.preprocessing import StandardScaler
dt_no_time['Amount'] = StandardScaler().fit_transform(dt_no_time['Amount'].values.reshape(-1,1))
dt_no_time.head()
```

```python
# phân chia dữ liệu train test
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(dt_no_time, test_size = 0.2, random_state = 42)
 
# xây dựng lại dữ liệu train
X_train = X_train[X_train['Class']==0]
X_train = X_train.drop(columns=['Class'])

# xây dựng lại dữ liệu test 
Y_test = X_test['Class']
X_test = X_test.drop(columns=['Class'])

X_train = X_train.values
X_test = X_test.values



```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from keras.engine import input_layer
from keras.layers import Input, Dense
from keras.models import load_model, Model
from keras.callbacks import ModelCheckpoint

# Xây dựng model
input_dimension = X_train.shape[1]
hidden_size = 128 # giá trị tùy ý

input_layer = Input(shape=(input_dimension,))

# Encoder
encoder = Dense(hidden_size, activation="relu") (input_layer)
encoder = Dense(hidden_size // 2, activation="relu")(encoder)
encoder = Dense(hidden_size // 4, activation="relu")(encoder)

#Decoder
decoder = Dense(hidden_size // 4, activation="relu")(encoder)
decoder = Dense(hidden_size // 2, activation="relu")(decoder)
decoder = Dense(hidden_size, activation="relu") (decoder)

# Output
output_layer = Dense(input_dimension, activation = "relu") (decoder)
auto_encoder_model = Model(inputs = input_layer, outputs = output_layer)
auto_encoder_model.compile(optimizer = "adam", loss = "mse", metrics=['accuracy'])
auto_encoder_model.summary()



```

```python
# Train model
n_epochs = 50
n_batch_size = 32

save_best = ModelCheckpoint(filepath = "/content/gdrive/MyDrive/data/best.h5", 
                            monitor='val_loss', verbose=1, save_best_only=True)
history = auto_encoder_model.fit(X_train, X_train, batch_size=n_batch_size, epochs=n_epochs, verbose = 1,
                                shuffle = True, callbacks=[save_best], validation_data=(X_test, X_test))
```

```python
# In history
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label = "loss") 
plt.plot(history.history['val_loss'], label= 'Val_loss')
plt.legend()

plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()


```

```python
# kiểm tra việc tái tạo trong tập test
import numpy as np

# Dự đoán
predicts = auto_encoder_model.predict(X_test)
mse = np.mean(np.power(X_test - predicts, 2), axis = 1)
err_dt = pd.DataFrame({'error': mse, 'truth' : Y_test})
```

```python
# In ra histogram của error khi tái tạo các điểm Normal trong tập test
normal_error_dt = err_dt[(err_dt['truth']==0) & (err_dt['error'] < 10)]

_ = plt.hist(normal_error_dt.error.values, bins = 50)
```

```python
# In ra histogram cuar error khi tái tạo các điểm Fraud trong tập test
fraud_error_dt = err_dt[err_dt['truth'] == 1]
_ = plt.hist(fraud_error_dt.error.values, bins = 100)
```

```python
# Vẽ confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sb

threshold = 5

y_pred = [1 if e > threshold else 0 for e in err_dt.error.values]
conf_matrix = confusion_matrix(err_dt.truth, y_pred)

# Vẽ
sb.heatmap(conf_matrix, xticklabels=["Normal", "Fraud"], yticklabels=["Normal", "Fraud"], annot=True, fmt="d")

plt.title("Confusion Matrix")
plt.xlabel("Predited Values")
plt.ylabel("Truth")
plt.show
```
